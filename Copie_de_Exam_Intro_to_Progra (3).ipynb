#**Project idea:** Accordingly, I would like to investigate the evolution of the pandemic per continent since 2020 and which one handled it better.

#I) First step is to select the columns I need to perform my investigation:

```python
covid_data= covid_data[['continent', 'date', 'total_cases_per_million', 'total_deaths_per_million', 'total_vaccinations_per_hundred', 
                        'population', 'median_age', 'aged_70_older', 'hospital_beds_per_thousand', 'human_development_index']]
covid_data
```

#II) Now, I would like to clean the data and remove all missing values:

```python
#First, I remove all the data that miss one of this information
covid_data= covid_data.dropna(thresh=1, subset= ['continent', 'date', 'total_cases_per_million', 'total_deaths_per_million', 
                                                 'total_vaccinations_per_hundred', 'population', 'median_age', 'aged_70_older', 
                                                 'hospital_beds_per_thousand', 'human_development_index'])
covid_data
```

```python
#I check where data are missing
covid_data.isnull().sum()
```

```python
#Now, I remove the data that are missing in each of these columns
covid_data= covid_data.dropna(thresh=1, subset= ['continent'])
covid_data= covid_data.dropna(thresh=1, subset= ['total_cases_per_million'])
covid_data= covid_data.dropna(thresh=1, subset= ['total_deaths_per_million'])
covid_data= covid_data.dropna(thresh=1, subset= ['total_vaccinations_per_hundred'])
covid_data= covid_data.dropna(thresh=1, subset= ['population'])
covid_data= covid_data.dropna(thresh=1, subset= ['median_age'])
covid_data= covid_data.dropna(thresh=1, subset= ['aged_70_older'])
covid_data= covid_data.dropna(thresh=1, subset= ['hospital_beds_per_thousand'])
covid_data= covid_data.dropna(thresh=1, subset= ['human_development_index'])
```

```python
#I check if there are still missing values
covid_data.isnull().sum()
```

#III) I am modifying the dataset to obtain a better representation

```python
#For the later representation, I need to create a column that would indicate how well each continent handled the corona crisis:
covid_data['handled_corona'] = (covid_data['total_deaths_per_million']).astype(int) / (covid_data['total_cases_per_million']).astype(int)

covid_data.head()
```

#VI) Interactive visualization of the data:

```python
import ipywidgets as widgets
from ipywidgets import interact
from ipywidgets import fixed
```

```python
#I would like to observe which continent handled the best the corona crisis
def select_continent(continent, df):

  return df.loc[df['continent'] == continent, ]

interact(select_continent, continent= ['Africa','Asia', 'Europe', 'North America', 'Oceania', 'South America'], df=fixed(covid_data));
```

```python
def visualize_continent(continent, df):
  
  continent_data = select_continent(continent, df)

chart = alt.Chart(continent_data).mark_bar().encode(
              x='date',
              y= 'handled_corona',
              color= 'continent'
          )
 
display(chart)
```

```python
visualization_handled_corona= alt.Chart(covid_data).mark_bar().encode(
      x='date',
      y='handled_corona',
      color='continent'
  )

visualization_handled_corona
```
